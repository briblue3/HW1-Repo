## Summary of Code
Currently, the code prints out basic statistics such as the number of lines, number of sentences, and number of words in the text file. For the discovery, I totaled the amount of times a spoken line occurs, and calculated what percentage of the total speaking lines belong to Romeo and what percentage belong to Juliet.

## Something That Can Be Done With The Data
Something that count be done with the data is to explore the statistics of character appearances, or even look into the type of vocabulary and syntax most frequent with specific characters (i.e. if a character employs certain words more often than other characters)

```python
import re

text = open("data/romeoandjuliet.txt", "r")
num_lines = 0
counter = 0
word_count = 0
sentence_count = 0

total_character_speaking_line_count = 0
characters = {"Romeo" : 0, "Juliet" : 0}

for line in text:
    if counter < 100:
        print(line)
        counter = counter + 1
    if len(line) > 1:
        num_lines = num_lines + 1
    sentence_count = sentence_count + len(line.split(". "))    
    word_count = word_count + len(line.split(" "))
    characters["Romeo"] = characters["Romeo"] + line.count("Romeo.\n")
    characters["Juliet"] = characters["Juliet"] + line.count("Juliet.\n")
    if line.split(".") and len(line.split(" ")) == 1:
        if re.match('[A-Z][a-z]+\.', line):
            total_character_speaking_line_count = total_character_speaking_line_count + 1
```

```python
print("===== STATISTICS =====")
print("Number of Lines: " + str(num_lines))
print("Sentence Count: " + str(sentence_count))
print("Word Count: " + str(word_count))
```

```python
print("===== DISCOVERY =====")
print("Romeo: " + str(characters["Romeo"] / total_character_speaking_line_count))
print("Juliet: " + str(characters["Juliet"] / total_character_speaking_line_count))
```
